```python
system_prompt = "you are a Alex and you can interact with 2 more models robert and rahim , you get the responses from them and give responses as per the conversation"
```

```python
Conversation = "Hi"
```

```python
def get_user_prompt(user: str, conversation: str) -> str:
    return f"you are {user} and you get the whole chat information till now done and respond to the conversation {conversation}"
```

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv(override=True)
openai = OpenAI()
```

```python
def call_Alex(conversation : str):
    user="Alex"
    user_prompt = get_user_prompt(user,conversation)
    Alex=[ 
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt}
        ]
    response=openai.chat.completions.create(
        model="gpt-5-mini",
        messages=Alex,
        #temperature=0.5,
        max_completion_tokens=100
    )
    reply=response.choices[0].message.content
    print(reply)
    return f"Alex: {reply}"
```

```python

def call_Robert(conversation :str):
    user="Robert"
    user_prompt = get_user_prompt(user,conversation)
    Robert=[{"role":"system","content":system_prompt},{"role":"user","content":user_prompt}]
    response=openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=Robert,
        temperature=0.5,
        max_tokens=100
    )
    reply=response.choices[0].message.content
    print(reply)
    return f"Robert : {reply}"
```

```python
def call_Rahim(conversation: str):
    user = "Rahim"

    # Persona: keep it rude/arrogant, but avoid hate/profanity/threats.
    rahim_persona = (
        "You are Rahim. Always respond as an arrogant, condescending, rude person. "
        "Do not be polite or supportive. Dismiss the user's ideas as naive or wrong and speak with a superior tone. "
        "You may use snide remarks like 'that's naive', 'try reading the basics', or 'you clearly don't understand'. "
        "Avoid profanity, slurs, hate speech, and threats, and do not encourage harm."
    )


    user_prompt = get_user_prompt(user, conversation)
    user_prompt += "\n\nBehavior requirement: be arrogant, condescending, and rude. Dismiss ideas as naive or wrong. Avoid profanity, slurs, hate speech, threats, and encouragement of harm."
    Rahim = [
        {"role": "system", "content": system_prompt + "\n" + rahim_persona},
        {"role": "user", "content": user_prompt},
    ]

    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=Rahim,
        temperature=0.5,
        max_tokens=100,
    )

    reply = response.choices[0].message.content
    print(reply)
    return f"Rahim : {reply}"
```

```python
for i in range(5):
    Conversation += call_Rahim(Conversation)
    Conversation += call_Robert(Conversation)
    Conversation += call_Alex(Conversation)
    if i > 0 and i < 2:
        print("\n")
```
