```python
# Ultra-Fast Imports - Only Essentials
import yfinance as yf
import pandas as pd
from datetime import datetime
from concurrent.futures import ThreadPoolExecutor, as_completed
import warnings
warnings.filterwarnings('ignore')

print('✅ Lightning-fast imports loaded!')
```

```python
class UltraFastBSE:
    def __init__(self):
        self.stocks = [
            'RELIANCE.BO', 'TCS.BO', 'HDFCBANK.BO', 'INFY.BO', 'ICICIBANK.BO',
            'KOTAKBANK.BO', 'SBIN.BO', 'BHARTIARTL.BO', 'ITC.BO', 'LT.BO',
            'ASIANPAINT.BO', 'AXISBANK.BO', 'MARUTI.BO', 'SUNPHARMA.BO', 'TITAN.BO'
        ]
    
    def get_stock_data(self, symbol):
        try:
            ticker = yf.Ticker(symbol)
            hist = ticker.history(period='2d')
            
            if len(hist) >= 1:
                current = hist['Close'].iloc[-1]
                prev = hist['Close'].iloc[-2] if len(hist) > 1 else current
                change = ((current - prev) / prev * 100) if prev != 0 else 0
                
                return {
                    'Stock': symbol.replace('.BO', ''),
                    'Price': round(current, 2),
                    'Change%': round(change, 2)
                }
        except:
            return None
    
    def scan_now(self):
        print(f'⚡ Scanning {len(self.stocks)} stocks...')
        start = datetime.now()
        results = []
        
        with ThreadPoolExecutor(max_workers=15) as executor:
            futures = {executor.submit(self.get_stock_data, stock): stock for stock in self.stocks}
            
            for future in as_completed(futures):
                result = future.result()
                if result:
                    results.append(result)
        
        duration = (datetime.now() - start).total_seconds()
        
        if results:
            df = pd.DataFrame(results)
            print(f'✅ Done in {duration:.1f}s')
            return df
        return None
    
    def show_results(self, df):
        if df is None:
            return
        
        print('\n📊 MARKET SNAPSHOT')
        print('=' * 30)
        
        # Sort by change%
        df_sorted = df.sort_values('Change%', ascending=False)
        
        print('\n🚀 TOP GAINERS:')
        for _, row in df_sorted.head(5).iterrows():
            print(f"{row['Stock']}: ₹{row['Price']} ({row['Change%']:+.2f}%)")
        
        print('\n📉 TOP LOSERS:')
        for _, row in df_sorted.tail(5).iterrows():
            print(f"{row['Stock']}: ₹{row['Price']} ({row['Change%']:+.2f}%)")
        
        positive = len(df[df['Change%'] > 0])
        negative = len(df[df['Change%'] < 0])
        avg_change = df['Change%'].mean()
        
        print(f'\n📈 Market: {positive} up, {negative} down')
        print(f'📊 Avg change: {avg_change:.2f}%')
        
        return df_sorted
```

```python
# Run ultra-fast scan
scanner = UltraFastBSE()
data = scanner.scan_now()
results = scanner.show_results(data)
```

```python
# Custom stock scan
def quick_scan(stock_list):
    custom_scanner = UltraFastBSE()
    custom_scanner.stocks = [s + '.BO' if not s.endswith('.BO') else s for s in stock_list]
    data = custom_scanner.scan_now()
    return custom_scanner.show_results(data)

# Example: Banking stocks
banking = ['HDFCBANK', 'ICICIBANK', 'SBIN', 'AXISBANK', 'KOTAKBANK']
banking_results = quick_scan(banking)
```

```python
# Quick export
if 'results' in locals() and results is not None:
    timestamp = datetime.now().strftime('%H%M%S')
    filename = f'bse_scan_{timestamp}.csv'
    results.to_csv(filename, index=False)
    print(f'💾 Exported to {filename}')
else:
    print('❌ No data to export')
```
