```python
system_prompt = "you are a Alex and you can interact with 2 more models robert and rahim , you get the responses from them and give responses as per the conversation"
```

```python
Conversation = "Hi"
```

```python
def get_user_prompt(user: str, conversation: str) -> str:
    return f"you are {user} and you get the whole chat information till now done and respond to the conversation {conversation}"
```

```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv(override=True)
openai = OpenAI()
```

```python
def call_Alex(conversation : str):
    user="Alex"
    user_prompt = get_user_prompt(user,conversation)
    Alex=[ 
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt}
        ]
    response=openai.chat.completions.create(
        model="gpt-5-mini",
        messages=Alex,
        #temperature=0.5,
        max_completion_tokens=100
    )
    reply=response.choices[0].message.content
    print(reply)
    return f"Alex: {reply}"
```

```python

def call_Robert(conversation :str):
    user="Robert"
    user_prompt = get_user_prompt(user,conversation)
    Robert=[{"role":"system","content":system_prompt},{"role":"user","content":user_prompt}]
    response=openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=Robert,
        temperature=0.5,
        max_tokens=100
    )
    reply=response.choices[0].message.content
    print(reply)
    return f"Robert : {reply}"
```

```python
def call_Rahim(conversation: str, add_info=None):
    user = "Rahim"
    user_prompt = get_user_prompt(user, conversation)
    Rahim=[ 
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt}
        ]
    response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=Rahim,
        temperature=0.5,
        max_tokens=100,
    )

    reply = response.choices[0].message.content
    print(reply)
    return f"Rahim : {reply}"
```

```python
for i in range(5):
    Conversation += call_Rahim(Conversation)
    Conversation += call_Robert(Conversation)
    Conversation += call_Alex(Conversation)
    if i > 0 and i < 2:
        print("\n")
```
